```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```python
wine = pd.read_csv(url, delimiter=';')
```

```python
wine.rename(columns = {"fixed acidity": "fixed_acidity",
                       "volatile acidity": "volatile_acidity",
                       "citric acid": "citric_acid",
                       "residual sugar": "residual_sugar",
                       "chlorides": "chlorides",
                       "free sulfur dioxide": "free_sulfur_dioxide",
                       "total sulfur dioxide": "total_sulfur_dioxide"},
            inplace = True)
```

```python
wine.head()
```

```python
columns = list(wine.columns)

print(f"New Columns: {columns}")
```

```python
sns.countplot(x='quality', data=wine)
plt.title('Wine Quality Distribution')
plt.show()
```

```python
sns.histplot(wine['alcohol'], bins=25, kde=True, color='purple')
plt.title('Alcohol Distribution')
plt.xlabel('Alcohol Content (% vol)')
plt.show()
```

```python
plt.subplot(1, 2, 1)
sns.histplot(wine['alcohol'], bins=25, kde=True, color='purple')
plt.title('Alcohol Distribution')
plt.xlabel('Alcohol Content (% vol)')

plt.subplot(1, 2, 2)
sns.boxplot(x=wine['alcohol'], color='orange')
plt.title('Alcohol Content Spread')
plt.xlabel('Alcohol (% vol)')

plt.show()
```

```python
sns.histplot(wine['pH'], bins=30)
plt.title('pH Distribution')
plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=wine)
plt.title('Alcohol Content by Quality Rating')
plt.show()
```

```python
quality_divided = wine['quality_tier'] = pd.cut(wine['quality'],
                             bins=[0, 5, 6, 10],
                             labels=['Low', 'Medium', 'High'])
```

```python
sns.countplot(x = quality_divided)
plt.show()
```

```python
sns.countplot(x='quality_tier', data=wine)
plt.title('Wine Quality Tier Distribution')
plt.xlabel('Quality Tier')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='quality', y='sulphates', data=wine)
plt.title('Sulphates Distribution by Quality Category')
plt.show()
```
