### **What Is Seaborn in Python?**

Python Seaborn library is a widely popular data visualization library that is commonly used for data science and machine learning tasks. You build it on top of the matplotlib data visualization library and can perform exploratory analysis. You can create interactive plots to answer questions about your data.

```python
sns.barplot(x='day', y='total_bill', data=df_1)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df_1)

plt.title('Average')
plt.xlabel('Days')
plt.ylabel('Total Bill')

plt.grid()
plt.show()
```

```python
time_counts = df_1['time'].value_counts()

sns.countplot(x='time', data=df_1)
plt.title('Number of Customers at Lunch and Dinner')
plt.xlabel('Time of Day')
plt.ylabel('Number of Customers')
plt.show()
```

```python
sns.countplot(x = 'sex', data=df_1)
plt.show()
```

```python
sns.histplot(df_1['total_bill'])
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df_1)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df_1, alpha=0.7)

plt.title('Total Bill vs. Tip Amount', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip Amount ($)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
