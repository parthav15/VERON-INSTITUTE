**Check is there any null values or missing values in our dataset**

```python
Book_Data.isnull().sum()

# There are  null values in the dataset, hence it need to be cleaned.
```

```python
Book_Data.head()
```

```python
Book_Data['Publishing Year'] = Book_Data['Publishing Year'].fillna(Book_Data['Publishing Year'].mode()[0])
```

```python
Book_Data['Book Name'] = Book_Data['Book Name'].fillna(Book_Data['Book Name'].mode()[0])
```

```python
Book_Data['language_code'] = Book_Data['language_code'].fillna(Book_Data['language_code'].mode()[0])
```

```python
Book_Data.isnull().sum()
```

```python
Book_Data['genre'].value_counts()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="genre", data=Book_Data)
plt.xticks(rotation='vertical')
plt.title('Book Mosts Release',fontsize=40)
plt.ylabel('Count',fontsize=25)
plt.xlabel('genre',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(Book_Data['Book_ratings_count'], bins=20, kde=True)
plt.title('Distribution of Book Ratings Count', fontsize=18)
plt.xlabel('Number of Ratings', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

***This visualization will analyze that how the number of ratings per book is distributed in given dataset.***

**Top 10 Authors with Most Books**

```python
Top_authors = Book_Data['Author'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=Top_authors.index, y=Top_authors.values)
plt.title('Top 10 Authors with Most Books Published', fontsize=18)
plt.xlabel('Author', fontsize=14)
plt.ylabel('Number of Books Published', fontsize=14, labelpad=20, verticalalignment='center')
plt.xticks(rotation=45,ha='center')
plt.show()
```

***From above diagram Stephen King is most renowned author with most publishing book and Lemony Snicket has less book published***
