- age: 177 missing values

- embarked: 2 missing values

- deck: 688 missing values

- embark_town: 2 missing values

```python
median_age = titanic['age'].median()

median_age
```

```python
titanic.fillna({'age': median_age}, inplace = True)
```

```python
titanic.head()
```

```python
mode_embarked = titanic['embarked'].mode()[0]
mode_embarked
```

```python
titanic.fillna({'embarked': mode_embarked}, inplace = True)
```

```python
titanic.isnull().sum()
```

```python
titanic['deck'] = titanic['deck'].astype(str)

titanic.fillna({'deck': 'Unknown'}, inplace = True)
```

```python
titanic.drop(['who', 'embark_town', 'alive'], axis=1, inplace=True)
```

```python
titanic.head()
```

### **SURVIVAL ANALYSIS**

```python
sns.countplot(x = 'survived', data = titanic)
plt.title('Survival Distribution (0 = Died, 1 = Survived)')
plt.show()
```

```python
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='survived', hue='class', data=titanic)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(titanic['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=titanic)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
titanic['age_group'] = pd.cut(titanic['age'],
                              bins=[0, 12, 19, 65, 100],
                             labels=['Child', 'Teen', 'Adult', 'Elderly'])

sns.countplot(x='age_group', hue='survived', data=titanic)
plt.title('Survival by Age Groups')
plt.show()
```

```python
sns.histplot(titanic['fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
titanic['class'].value_counts()
```

```python
titanic['class'].value_counts().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=titanic)
plt.title('Fare Distribution by Class')
plt.show()
```

```python
sns.countplot(x='embarked', data=titanic)
plt.title('Passengers by Embarkation Port')
plt.show()
```

```python
sns.barplot(x='embarked', y='survived', data=titanic)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```
